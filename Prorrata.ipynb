```python

# obtengo la data de los resultados del codigo anterior
Carpeta = "Resultados"
excel = "Data_SEN_target.xlsx"
ruta = Carpeta + "/" +excel
df_concatenado= pd.read_excel(ruta,sheet_name="Data")

#--------------------------------------------------------------------
df_nuevo=df_concatenado[["Name1","Name2","EMTP Status","Type","Nombre PowerFactory","Página PowerFactory","Nº_unid","Nº_unid_serv"]]

df_concatenado['Nº_unid'] = df_concatenado['Nº_unid'].astype(float)
df_concatenado['Nº_unid_serv'] = df_concatenado['Nº_unid_serv'].astype(float)
#df_nuevo['Potencia Nominal [MVA]'] = df_nuevo['Potencia Nominal [MVA]'].astype(float)
df_nuevo["Snom/unidad [MVA]"] = df_concatenado["Potencia Nominal [MVA]"]/df_concatenado["Nº_unid"]


#df_nuevo["Despacho/unidad [pu]"]=df_concatenado["P despachada [MW]"]/df_concatenado["Sbase [MVA]"] 
#df_nuevo["Q/unidad [pu]"] =df_concatenado["P despachada [MW]"]/df_concatenado["Sbase [MVA]"] 
df_nuevo["Despacho/unidad [pu]"]=df_concatenado["P_set [MW]"]/df_concatenado["Sbase [MVA]"] 
df_nuevo["Q/unidad [pu]"] =df_concatenado["Q_set [Mvar]"]/df_concatenado["Sbase [MVA]"] 

Data_restructurada= df_nuevo

#----------------------------------------------------------------------------------------
# Filtro las PFV Y PE que estan en una zona determinada

tipo = ['Parque Fotovoltaico', 'Parque Eólico',"PMGD"]
data = Data_restructurada[Data_restructurada['Type'].isin(tipo)]


# filtro las zonas necesarias 
paginas = ['00-Norte Grande', '01-Atacama','02-Coquimbo','03-Chilquinta-Aconcagua']

data = data[data['Página PowerFactory'].isin(paginas)]
df_filtrado = data[data['EMTP Status'] != 0]

df_nuevo
```

```python

#----------------------- Agrego columna que define si esta o no arriba de polpaico-------------------------------------------------
c=-1



for Name1 in df_nuevo["Name1"]:
    c=c+1
    if Name1 in list(df_filtrado["Name1"]):
        df_nuevo.at[c,'Arriba Polpaico']=1
    else : 
        df_nuevo.at[c,"Arriba Polpaico"]=0   

pd.set_option('display.max_rows', 351)
df_nuevo 
```

```python
#--------- Configuracion adicional 
df_nuevo['Despacho/unidad [pu]'] = df_nuevo['Despacho/unidad [pu]'].astype(float)
df_nuevo['Q/unidad [pu]'] = df_nuevo['Q/unidad [pu]'].astype(float)
df_nuevo = df_nuevo.rename(columns={'Nº_unid_serv':'N°unidades en servicio'})
df_nuevo = df_nuevo.rename(columns={'Nº_unid':'N° unidades totales'})


df_nuevo['Despacho/unidad [pu]'] = df_nuevo['Despacho/unidad [pu]'].round(3)
df_nuevo['Q/unidad [pu]'] = df_nuevo['Q/unidad [pu]'].round(4)
df_nuevo['N° unidades totales'] = df_nuevo['N° unidades totales'].astype(float)
df_nuevo['N°unidades en servicio'] = df_nuevo['N°unidades en servicio'].astype(float)
pd.set_option('display.max_rows', 320)
df_nuevo   
df_nuevo   
```
